```python
#Q2.Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
#release) and make data frame
```

```python
import  requests
from  bs4 import BeautifulSoup
url = requests.get('https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating')
url
```

```python

soup = BeautifulSoup(url.text, 'html.parser')
```

```python

movie_name=[]
rating=[]
year=[]
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
    year.append(year_of_release)
    
    
    rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
    rating.append(rate)
    
```

```python
movie_name
```

```python
rating
```

```python
year
```

```python
#Making as a DataFrame

import pandas as pd
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of relase': year, 'Movie Rating': rating})
movie_DF
```

```python
#Q8.Write a python program to scrape mentioned details from ‘https://www.dineout.co.in/delhi-restaurants/buffet- special’ :
#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image url
import  requests
import pandas as pd
from  bs4 import BeautifulSoup
url = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
url
```

```python
soup = BeautifulSoup(url.text, 'html.parser')
```

```python
# Finding the names of restaurants 

Resturant_Name=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    Resturant_Name.append(i.text)

Resturant_Name
```

```python
#Finding names of cuisines 

cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])

cuisine
```

```python
# Find the location
location=[]
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    location.append(i.text)
location
```

```python
# Find the ratings
ratings=[]
for i in soup.find_all('div',class_='img-wrap'):
    ratings.append(i.text)
ratings
```

```python
# find the imageurl
image_url=[]
for i in soup.find_all('img',class_='no-img'):
    image_url.append(i['data-src'])
image_url
```

```python
# chking length
print(len(Resturant_Name),len(cuisine),len(location),len(ratings),len(image_url))
```

```python
#make as a DataFrame
import pandas as pd
df=pd.DataFrame({'Resturant_Name':Resturant_Name,'location':location,'cuisine':cuisine,'Ratings':ratings,'Image_Url':image_url})
df
```

```python
#Q9Write a python program to scrape weather details for last 24 hours from
#‘https://en.tutiempo.net/delhi.html?data=last-24-
#i) Hour
#ii) Temperature
#iii) Wind
#iv) Weather condition v) Humidity
#vi) Pressure

import  requests
import pandas as pd

from  bs4 import BeautifulSoup
url = requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')
url
```

```python
# Weather condition
soup = BeautifulSoup(url.text, 'html.parser')
Weather_condition=[]
for i in soup.find_all ('td',class_='t Temp')[0:24]: 

    if i.previous_sibling is not None:

        Weather_condition.append(i.previous_sibling.text)

    else:

        Weather_condition.append(' ')
Weather_condition

```

```python
# WIND
wind=[]
for i in soup.find_all ('td',class_='hr')[0:24]: 

    if i.previous_sibling is not None:

        wind.append(i.previous_sibling.text)

    else:

        wind.append(' ')
wind


```

```python
# TEMPERATURE
Temp=[]
for i in soup.find_all ('td',class_='wind')[0:24]: 

    if i.previous_sibling is not None:

        Temp.append(i.previous_sibling.text)

    else:

        Temp.append(' ')
Temp
```

```python
# humdity
hum=[]
for i in soup.find_all ('td',class_='prob')[0:24]: 

    if i.previous_sibling is not None:

        hum.append(i.previous_sibling.text)

    else:

        hum.append(' ')
hum
```

```python
#PRESSURE
pre=[]
for i in soup.find_all ('td',class_='prob')[0:24]:
    pre.append(i.text)
pre
    
```

```python
# HOUR
Hour=[]
for i in soup.find_all ('td',class_='t Temp')[0:24]: 

    if i.previous_sibling.previous_sibling is not None:

        Hour.append(i.previous_sibling.previous_sibling.text)

    else:

        Hour.append(' ')
Hour
```

```python
import pandas as pd
df=pd.DataFrame({'Hour':Hour,'Temp':Temp,'Wind':wind,'Pressure':pre,'Humidity':hum})
df
```
