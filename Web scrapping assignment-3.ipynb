# Q1.Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
class AmazonProd(object):                      #initiating class
    
    def __init__(self,items):                 #defining class object
        self.amazon_url="https://www.amazon.in/"   #passing url
        self.items = items                         #defining variables
        self.profile=webdriver.Chrome(r"/Users/mj/Downloads/chromedriver")
        self.driver=webdriver.Chrome("chromedriver")            #getting webdriver 
        self.driver.get(self.amazon_url)  
        
        
    def search_items(self):                          #defining new function with class object
        urls=[]
        prices=[]                                    #empty list to store data
        names=[]
        for item in self.items:                       #using for loop
            print(f"Searching for {items}:")
            
            self.driver.get(self.amazon_url)
            search_field=self.driver.find_element_by_id("twotabsearchtextbox")    #fetching search tag using id name
            search_field.send_keys(item)                                      #sending keys into search field
            
            time.sleep(2)
            
            search_button=self.driver.find_element_by_id("nav-search-submit-button")  #fetching search tag using id name
            search_button.click()                                      #submitting search_button
            
            time.sleep(1)
items=[input("Enter item")]                                 # making search_field as user input
try:
    amazon_pro=AmazonProd(items)
    amazon_pro.search_items()                               #using try and except method
    print("Item searched")                                 
except:
    print("No Such item exist")
```

# Q6:Write a program to scrap details of all the funding deals for second quarter(i.e.July 20 - september 20) from trak.in

```python
#importing webdriver
driver=webdriver.Chrome(r"/Users/mj/Downloads/chromedriver")
time.sleep(4)
url="https://trak.in/india-startup-funding-investment-2015/"
driver.get(url)  #getting the funding page
time.sleep(4)

#creating empty boxex
date=[]
s_name=[]
industry=[]
sub=[]
location=[]
i_name=[]
i_type=[]
amount=[]
#getting the data of the month from july to september
for i in range(48,51):
    # Date
    dt = driver.find_elements_by_xpath('//table[@id="tablepress-{}"]/tbody/tr/td[2]'.format(i))
    for d in dt:
        date.append(d.text)

    #startupname
    s=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[3]'.format(i))
    for d in s:
        s_name.append(d.text)

    #industry
    a=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[4]'.format(i))
    for d in a:
        industry.append(d.text)
    #sub
    su=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[5]'.format(i))
    for d in su:
        sub.append(d.text)
    #location
    lo=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[6]'.format(i))
    for d in lo:
        location.append(d.text)
    #i_name
    ina=driver.find_elements_by_xpath("//*[@id='tablepress-{}']/tbody/tr/td[7]".format(i))
    for d in ina:
        i_name.append(d.text)
    #i_type
    ty=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[8]'.format(i))
    for d in ty:
        i_type.append(d.text)
    #amount
    am=driver.find_elements_by_xpath('//*[@id="tablepress-{}"]/tbody/tr/td[9]'.format(i))
    for d in am:
        amount.append(d.text)

time.sleep(2)
#creating dataframe
ans6=pd.DataFrame({})
ans6["Date"]=date
ans6["Startup_Name"]=s_name
ans6["Industry"]=industry
ans6["Sub"]=sub
ans6["Location"]=location
ans6["Investors name"]=i_name
ans6["Investment_type"]=i_type
ans6["Amount"]=amount
ans6
```

# Q9:Write a program to extract atleast 500 comments, comment upvote and time when comment was posted from any Youtube video.

```python
driver = webdriver.Chrome(r"/Users/mj/Downloads/chromedriver")
#get the web page with given url
url = "https://www.youtube.com/"
driver.get(url)
time.sleep(3)

#lets click on a video
clk_video = driver.find_element_by_xpath("/html/body/ytd-app/div/ytd-page-manager/ytd-browse/ytd-two-column-browse-results-renderer/div[1]/ytd-rich-grid-renderer/div[6]/ytd-rich-item-renderer[3]/div/ytd-rich-grid-media/div[1]/div/div[1]/h3/a").click()
time.sleep(8)

#Lets scroll the comments

for _ in range(1000):

    driver.execute_script("window.scrollBy(0,100)")

#fetching required data
comments = []
upvotes = []
times = []
try:
    comment = driver.find_elements_by_xpath("//*[@id='content-text']")[:500]
    for i in comment:
        comments.append(i.text)
except NoSuchElementException:
    comments.append('-')
time.sleep(3)

#fetching upvotes
try:
    up = driver.find_elements_by_xpath("//*[@id='vote-count-middle']")[:500]
    for i in up:
        upvotes.append(i.text)
except NoSuchElementException:
    comments.append('-')
    time.sleep(3)

#fetchin time when comment was posted
try:
    tim = driver.find_elements_by_xpath("//yt-formatted-string[@class='published-time-text above-comment style-scope ytd-comment-renderer']")[:500]
    for i in tim:
        times.append(i.text.replace('(edited)',''))
except NoSuchElementException:
    times.append('-')
time.sleep(3)

#creating dataframe
data = list(zip(comments,upvotes,times))
df = pd.DataFrame(data, columns = ["Comment","Upvotes","Published time"])
df
```

```python
import selenium
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
driver = webdriver.Chrome(r"/Users/mj/Downloads/chromedriver")
# obtaining the web page from the given url
url = "https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-07-23&to=2021-07-26&guests"
driver.get(url)
```

```python
Hostel_Name1 = driver.find_elements_by_xpath("//h2[@class='title title-6']/a")
Hostel_Name = []
for i in Hostel_Name1:
    Hostel_Name.append(i.text)        
```

```python
Distance_from_City_Centre = []
```

```python
rating = driver.find_elements_by_xpath("//div[@class='score orange big']")
```

```python
Ratings = []
for i  in rating:
    try:
        Ratings.append(i.text)
    except NoSuchElementException as E:
        Ratings.append("__")
```

```python
Total_Reviews1 = driver.find_elements_by_xpath("//div[@class='reviews']")
```

```python
Total_Reviews =[]
for i in Total_Reviews1:
    Total_Reviews.append(i.text)
```

```python
Overall_reviews1 = driver.find_elements_by_xpath("//div[@class='keyword']")
```

```python
Overall_Reviews =[]
for i in Overall_reviews1:
    Overall_Reviews.append(i.text)
```

```python
privates_from_price1 =driver.find_elements_by_xpath("//div[@class='price title-5']")
Privates_from_price =[]
for i in privates_from_price1:
    Privates_from_price.append(i.text)
```

```python
Dorms_from_price = Dorms_from_price[1::2]
```

```python
property_description1 = driver.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']")
Property_Description =[]
for i in property_description1:
    Property_Description.append(i.text)
```
